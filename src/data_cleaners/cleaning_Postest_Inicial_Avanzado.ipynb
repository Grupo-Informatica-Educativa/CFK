## Leer la data, filtrar solo por la Encuesta correspondiente al postest.

```python
df_inicial = pd.read_csv('../../data/crudos/bd_full_inicial.csv',
                         error_bad_lines=False, warn_bad_lines=False,
                         low_memory=False)
df_avanzado = pd.read_csv('../../data/crudos/bd_full_avanzado.csv',
                          error_bad_lines=False, warn_bad_lines=False,
                          low_memory=False)

df_inicial = df_inicial[df_inicial.Encuesta == 'U4 Cuestionario Final (25 puntos)']
df_avanzado = df_avanzado[df_avanzado.Encuesta == 'UA4 Cuestionario final (15 puntos)']
```

```python
# PIVOT DATA
pivot_inicial = df_inicial.pivot_table(
    index=['Nombre', 'Apellido', 'Correo Electrónico', 'Curso', 'ID Asignado Por Moodle', 'Nombre De Usuario'],
    columns='Pregunta', 
    values='Respuesta', 
    aggfunc='first').reset_index()

pivot_avanzado = df_avanzado.pivot_table(
    index=['Nombre', 'Apellido', 'Correo Electrónico', 'Curso', 'ID Asignado Por Moodle', 'Nombre De Usuario'],
    columns='Pregunta', 
    values='Respuesta', 
    aggfunc='first').reset_index()
```

```python
# Remove \n in columns
pivot_inicial.columns = [col.replace("\n", " ").strip() 
                        for col in pivot_inicial.columns]
pivot_avanzado.columns = [col.replace("\n", " ").strip() 
                        for col in pivot_avanzado.columns]
```
