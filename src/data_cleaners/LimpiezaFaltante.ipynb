### Dividir en subpreguntas la pregunta 22:

```python
opciones = ['Si, los estudiantes est√°n aprendiendo a recolectar, organizar, y analizar datos.',
'Si, los estudiantes est√°n aprendiendo sobre modelos computacionales para la predicci√≥n del clima',
'No, los estudiantes s√≥lo est√°n aprendiendo a cargar cosas en un archivo compartido y crear un gr√°fico',
'No, solo est√° usando computadores',
'!other=Otra:']
```

```python
col = '22. La docente Margarita decidi√≥ hacer que sus estudiantes de segundo de primaria utilicen los computadores del colegio para predecir el clima de una semana (temperatura, precipitaciones, y viento). Cada estudiante debe dibujar c√≥mo se ver√° el clima en la ciudad en dicha semana. Margarita, cre√≥ un archivo compartido donde los estudiantes ingresar√°n la informaci√≥n. Luego tomaron las predicciones de modelos de Internet y los ingresaron en el mismo documento compartido. Durante una semana tomaron los datos reales, y luego, proyectaron en el tablero los datos predichos por los estudiantes, los del modelo de Internet, y los datos reales. Al finalizar, Margarita les mostr√≥ a los estudiantes c√≥mo hacer un gr√°fico para comparar los diferentes datos. ¬øEst√° Margarita desarrollando el pensamiento computacional de sus estudiantes? Seleccione todas las respuestas que considere correctas.'
```

```python
def divide_question_YES_NO(num_col, str_col, opciones, df):
    for i, op in enumerate(opciones):
        print(num_col, i)
        subpreg = f'{num_col}.{i+1} {op}'
        mask = df[str_col].str.contains(op, na=False)
        df[subpreg] = mask
        df[subpreg].replace({True: 'SI', False: 'NO'}, inplace=True)
```

```python
divide_question_YES_NO(22, col, opciones, df_inicial)
df_inicial.drop(columns=col, inplace=True)
```
